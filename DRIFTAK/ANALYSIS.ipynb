```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
experiment_name = "lownoise"
n_runs = 4

actor_losses = []
critic_losses = []
ep_rewards = []
ep_mean_rewards = []

for run in range(n_runs):
    actor_losses.append(np.load(f"./experiments/{experiment_name}/actor_loss-{run}.npy"))
    critic_losses.append(np.load(f"./experiments/{experiment_name}/critic_loss-{run}.npy"))
    ep_rewards.append(np.load(f"./experiments/{experiment_name}/episodic_reward-{run}.npy"))
    ep_mean_rewards.append(np.load(f"./experiments/{experiment_name}/episodic_mean_reward-{run}.npy"))

ep_rewards = np.array(ep_rewards)
ep_mean_rewards = np.array(ep_mean_rewards)

t = np.arange(ep_rewards.shape[1])

plt.plot(t, ep_mean_rewards[3])

```
