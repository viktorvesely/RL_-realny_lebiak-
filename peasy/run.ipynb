```python
from simulation import simulateAll, summarise, simulateANDshow
from bandits import BanditType
from agent import ExplorationMethod
import matplotlib.pyplot as plt
```

```python
# experiment runs
N = 100
# training iterations
T = 1000
# arms
K = 10
```

```python
actions, rewards = simulateAll(N= N, T= T, k= K)
summarisedA, summarisedR = summarise(actions, rewards, N, T)
```

```python
plt.rcParams["figure.figsize"] = (25,5)

x = list(range(1,T+1))
plot1 = plt.figure(1)
fig, axs = plt.subplots(1, 2)
for key, value in summarisedA.items():
    if key[1] == BanditType.BERNOULLI:
        axs[0].plot(x, value, linewidth=0.5)
    else:
        axs[1].plot(x, value, label = key[0].name, linewidth=0.5)
axs[0].set_title("Percentage of optimal actions + Bernoulli distribution")
axs[1].set_title("Percentage of optimal actions + Normal distribution")
fig.legend()

plot2 = plt.figure(2)
fig, axs = plt.subplots(1, 2)
for key, value in summarisedR.items():
    if key[1] == BanditType.BERNOULLI:
        axs[0].plot(x, value, linewidth=0.5)
    else:
        axs[1].plot(x, value, label = key[0].name, linewidth=0.75)
axs[0].set_title("Average reward + Bernoulli distribution")
axs[1].set_title("Average reward + Normal distribution")
fig.legend()
```
